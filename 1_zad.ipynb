# Problem A1

```python
import math
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

#Laboratory 1
#Problem A1
#Random number generator -> normal distribution

def normal_distribution(x, standard_deviation, expected_value):
    return 1/(standard_deviation*(np.sqrt(np.pi*2)))*np.exp(-1/2*((x-expected_value)**2/(standard_deviation**2)))

def generate_random_numbers_polar(size):
    result = np.zeros(size)
    for i in range(size):
        radius = 1.1
        while radius > 1:
            u = np.random.uniform(-1, 1)
            v = np.random.uniform(-1, 1)
            radius = u**2 + v**2
        result[i] = u * np.sqrt(-2*np.log(radius)/radius)
    return result

x = np.linspace(-5,5, 10000)

y_generated = generate_random_numbers_polar(10000)

y_theoretical = normal_distribution(x, 1, 0)

plt.hist(y_generated, bins = 50, density=True)
plt.plot(x, y_theoretical)
```

```python
#Laboratory 1
#Problem A2
#Random number generator -> Cauchy's distribution

def cauchys_distribution(x, location, hwhm):
    return 1/(np.pi*hwhm)*(hwhm**2/((x-location)**2+hwhm**2))

def generate_random_numbers_exponential_inverse_trans(size, location, hwhm):
    U=np.random.uniform(0,1, size)
    return np.tan(np.pi*(U-1/2))*hwhm + location

#y_generated = exponential_inverse_trans(10000)

#plt.hist(y_generated, 50)

x = np.linspace(-5,5, 1000)
y_theoretical = cauchys_distribution(x, 0, 1)
y_generated = generate_random_numbers_exponential_inverse_trans(1000, 0, 1)
plt.plot(x, y_theoretical)
plt.hist(y_generated, 50, range=(-5,5), density=True)
```

# Problem B

```python
#Gambler's Ruin Problem

def a_ruin_probability(player_a_demesne, player_b_demesne, player_a_win_probability):
    u = ((1-player_a_win_probability)/(player_a_win_probability))**player_a_demesne
    v = ((1-player_a_win_probability)/(player_a_win_probability))**(player_a_demesne+player_b_demesne)
    return (u-v)/(1-v) if player_a_win_probability != 0.5 else 1 - player_a_demesne/(player_a_demesne+player_b_demesne)

def b_ruin_probability(player_a_demesne, player_b_demesne, player_b_win_probability):
    u = ((1-player_b_win_probability)/(player_b_win_probability))**player_b_demesne
    v = ((1-player_b_win_probability)/(player_b_win_probability))**(player_a_demesne+player_b_demesne)
    return (u-v)/(1-v) if player_b_win_probability != 0.5 else 1 - player_b_demesne/(player_a_demesne+player_b_demesne)

def players_ruin_game(player_a_initial, player_b_initial, player_a_win_probability, prediction_history_enabled=False):
    player_a_demesne = player_a_initial
    player_b_demesne = player_b_initial
    
    player_a_ruin_probability_history=[a_ruin_probability(player_a_demesne, player_b_demesne, player_a_win_probability)]
    player_b_ruin_probability_history=[b_ruin_probability(player_a_demesne, player_b_demesne, 1-player_a_win_probability)]
    
    #print(player_a_winning_probability_history[0])
    #print(player_a_winning_probability_history[0])
    
    round = 0
    while player_a_demesne and player_b_demesne:
        round+=1
        player_a_score = np.random.uniform(0,1)
        if player_a_score <= player_a_win_probability:
            player_a_demesne+=1
            player_b_demesne-=1       
        else:
            player_b_demesne+=1
            player_a_demesne-=1
        if prediction_history_enabled:
            player_a_ruin_probability_history.append(a_ruin_probability(player_a_demesne, player_b_demesne, player_a_win_probability))
            player_b_ruin_probability_history.append(b_ruin_probability(player_a_demesne, player_b_demesne, 1-player_a_win_probability))
    return {
        "winner": "a" if player_a_demesne else "b",
        "player_a_ruin_probability_history": player_a_ruin_probability_history,
        "player_b_ruin_probability_history": player_b_ruin_probability_history,
        "rounds_played": round
    }
    

#Problem B
def problem_b(player_a_initial_demesne, player_b_initial_demesne):
    number_of_games = 1000
    player_a_winning_probabilities = np.linspace(0.47, 0.53, 20) #Pick yourself
    #print(player_a_winning_probabilities)

    win_count = 0
    predicted_probability_sum = 0
    predicted_probability = 0

    game_results = []

    plotted_real_results = []
    plotted_predicted_results = []


    for player_a_winning_probability in player_a_winning_probabilities:
        for game in range(number_of_games):
            game_results.append(players_ruin_game(player_a_initial_demesne, player_b_initial_demesne, player_a_winning_probability))
        #print(game)
        for game in game_results:
            if game["winner"] == "a":
                win_count+=1
            predicted_probability+=game["player_a_ruin_probability_history"][0]

        #print(1-win_count/1000)
        plotted_real_results.append(1-win_count/1000)
        plotted_predicted_results.append(predicted_probability/1000)
        #print(predicted_probability/1000)
        predicted_probability_sum+=predicted_probability/1000
        win_count = 0
        predicted_probability = 0
        game_results.clear()

    print(predicted_probability_sum/20)
    plt.plot(player_a_winning_probabilities, plotted_real_results, '-o', label="Real Outcomes")
    plt.plot(player_a_winning_probabilities, plotted_predicted_results, '-o', label="Predicted values")
    plt.xlabel("Player a probability of round win")
    plt.ylabel("Player a ruin probability")
    plt.legend()
    plt.show()
```

```python
def players_ruin_game_times(player_a_initial, player_b_initial, player_a_win_probability, number_of_rounds):
    player_a_demesne = player_a_initial
    player_b_demesne = player_b_initial
    
    #print(player_a_winning_probability_history[0])
    #print(player_a_winning_probability_history[0])
    
    round = 0
    for i in range(number_of_rounds):
        round+=1
        player_a_score = np.random.uniform(0,1)
        if player_a_score <= player_a_win_probability:
            player_a_demesne+=1
            player_b_demesne-=1       
        else:
            player_b_demesne+=1
            player_a_demesne-=1
    return {
        "player_a_demesne":player_a_demesne,
        "player_b_demesne":player_b_demesne
    }

def problem_f(a_win_probability):
    number_of_rounds = [2, 10, 30, 50, 100] #edit manually
    a_win_probabilities = [a_win_probability]
    player_a_demesne = 50
    player_b_demesne = 50
    number_of_games = 5000
    player_a_capitals = []
    for idx, a_win_probability in enumerate(a_win_probabilities):
        for rounds in number_of_rounds:
            for game in range(number_of_games):
                player_a_capitals.append(players_ruin_game_times(player_a_demesne, player_b_demesne, a_win_probability, rounds)["player_a_demesne"])
            #plt.figure(100*idx+rounds)
            plt.hist(player_a_capitals, alpha=0.4)
            plt.xlabel("Player a demesne after number of rounds")
            plt.ylabel("Probability")
            player_a_capitals.clear()
    
```

# Problem G 

```python
def players_ruin_game_trajectory(player_a_initial, player_b_initial, player_a_win_probability):
    player_a_demesne = player_a_initial
    player_b_demesne = player_b_initial
    
    #print(player_a_winning_probability_history[0])
    #print(player_a_winning_probability_history[0])
    win_count = 0
    win_trajectory = [0]
    capital_trajectory = [player_a_initial]
    round = 0
    while player_a_demesne and player_b_demesne:
        round+=1
        player_a_score = np.random.uniform(0,1)
        if player_a_score <= player_a_win_probability:
            player_a_demesne+=1
            player_b_demesne-=1     
            win_count+=1
        else:
            player_b_demesne+=1
            player_a_demesne-=1
        capital_trajectory.append(player_a_demesne)
        win_trajectory.append(win_count)
    return (win_trajectory, capital_trajectory)

def problem_g():
    number_of_games = 10
    player_a_demesne = 50
    player_b_demesne = 50
    player_a_winning_probabilities = np.linspace(0.47, 0.53, 20)
    for idx, player_a_winning_probability in enumerate(player_a_winning_probabilities):
        for game in range(number_of_games):
            win_trajectory, capital_trajectory = players_ruin_game_trajectory(player_a_demesne,player_b_demesne,player_a_winning_probability)
            plt.figure(idx)
            plt.plot(win_trajectory)
            plt.figure(idx+1000)
            plt.plot(capital_trajectory)
```
